```python
URL = "https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD&bom=true&format=true&delimiter=%3B"
```

```python
from urllib.request import urlretrieve
urlretrieve(URL, 'Fremont.csv');
```

```python
import pandas as pd
data = pd.read_csv('Fremont.csv',  sep=';', parse_dates=True, index_col = 'Date')
data.head()
```

```python
%matplotlib inline
data.resample('W').sum().plot();
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
data.columns = ('West', 'East')
data.resample('W').sum().plot();
```

```python
data['Total'] = data['West']+data['East']
ax = data.resample('d').sum().rolling(365).sum().plot()
ax.set_ylim(0,None);
```

```python
data.groupby(data.index.time).mean().plot();
```

```python
pivoted = data.pivot_table('Total', index=data.index.time,columns=data.index.date)
pivoted.iloc[:5, :5]
```

```python
pivoted.plot(legend=False, alpha=0.02);
```
